# **Dropout Regularization: Preventing Overfitting in Neural Networks (MNIST)**


This tutorial explores how dropout regularization reduces overfitting in neural networks using the MNIST dataset.

```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical

from sklearn.metrics import confusion_matrix

plt.style.use("seaborn-v0_8-colorblind")
```

### Load dataset

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0


y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

print("Train shape:", X_train.shape)
```

### Build Model

```python
def build_model(dropout_rate=0.0):
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dropout(dropout_rate),
        Dense(64, activation='relu'),
        Dropout(dropout_rate),
        Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

### Train Models

```python
# No Dropout
model_no_dropout = build_model(0.0)
history_no = model_no_dropout.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)

# With Dropout
model_dropout = build_model(0.5)
history_drop = model_dropout.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)
```

### Accuracy Plot

```python
plt.figure(figsize=(8,5))

plt.plot(history_no.history['accuracy'], label='Train (No Dropout)')
plt.plot(history_no.history['val_accuracy'], label='Val (No Dropout)')

plt.plot(history_drop.history['accuracy'], linestyle='--', label='Train (Dropout)')
plt.plot(history_drop.history['val_accuracy'], linestyle='--', label='Val (Dropout)')

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Figure 1 – Dropout vs No Dropout Accuracy ")
plt.legend()

plt.savefig("figure1_accuracy.png", dpi=300)
plt.show()
```

### Loss plot

```python
plt.figure(figsize=(8,5))

plt.plot(history_no.history['loss'], label='Train (No Dropout)')
plt.plot(history_no.history['val_loss'], label='Val (No Dropout)')

plt.plot(history_drop.history['loss'], linestyle='--', label='Train (Dropout)')
plt.plot(history_drop.history['val_loss'], linestyle='--', label='Val (Dropout)')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Figure 2 – Dropout vs No Dropout Loss ")
plt.legend()

plt.savefig("figure2_loss.png", dpi=300)
plt.show()
```

### Different Dropout Rates

```python
rates = [0.2, 0.5]
histories = {}

for r in rates:
    model = build_model(r)
    histories[r] = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=8,
        batch_size=128,
        verbose=0
    )
```

### Compare Dropout rates

```python
plt.figure(figsize=(8,5))

for r in rates:
    plt.plot(histories[r].history['val_accuracy'], label=f'Dropout {r}')

plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.title("Figure 3 – Effect of Different Dropout Rates ")
plt.legend()

plt.savefig("figure3_dropout_rates.png", dpi=300)
plt.show()
```

### Overfitting Gap Plot

```python
gap_no = np.array(history_no.history['accuracy']) - np.array(history_no.history['val_accuracy'])
gap_drop = np.array(history_drop.history['accuracy']) - np.array(history_drop.history['val_accuracy'])

plt.figure(figsize=(8,5))

plt.plot(gap_no, label='Overfitting Gap (No Dropout)')
plt.plot(gap_drop, linestyle='--', label='Overfitting Gap (Dropout)')

plt.xlabel("Epochs")
plt.ylabel("Train - Validation Accuracy")
plt.title("Figure 4 – Overfitting Gap Comparison")
plt.legend()

plt.savefig("figure4_overfitting_gap.png", dpi=300)
plt.show()
```

### Confusion Matrix

```python
y_pred = model_dropout.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Figure 5 – Confusion Matrix (Dropout Model) ")

plt.savefig("figure5_confusion_matrix.png", dpi=300)
plt.show()
```

## Key Insights

- The model without dropout shows clear overfitting with higher training accuracy than validation accuracy.
- Dropout reduces overfitting by regularizing the network.
- The overfitting gap plot clearly shows improved generalization.
- Moderate dropout (0.2–0.5) gives better performance.
- The confusion matrix confirms strong classification performance.
